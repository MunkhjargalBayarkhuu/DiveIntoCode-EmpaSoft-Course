## [Problem 2] Create a function to calculate the gradient at a certain point

```python
x = Fuji[:, 0]
y = Fuji[:, 3]
```

```python
gradient_fuji = np.diff(y)/np.diff(x)
```

```python
plt.plot(gradient_fuji)
plt.title("Gradient of Mt.Fuji")
plt.show()
```

## [Problem 3] Create a function to calculate the destination point

```python
# y[t+1] = y[t]-alpha*gradient[t] #
```

```python
alpha = 0.2

def destination_point(index,
                              Mt_Fuji = Fuji[:, 3], 
                              gradient_mountain = gradient_fuji, 
                              alpha=0.2):

    '''
    Calculate go down the mountain
    
    Parameters
    -------------------------------
    Mt_Fuji:
        Current_position
    alpha:
        Descent change
        
     Returns
    -------------------------------
     y_t: int
        calculated destination point
    ''' 
   
    y_t = Mt_Fuji[index]-alpha*gradient_mountain[index]   
    
    return y_t
```

## [Problem 6] Change of initial value

```python
def get_grad_list(initial_point):
  
    gradient_list = []
    for index in range(initial_point, int(x[-1])):
        dist = destination_point(index=index)
        gradient_list.append(dist)

    return gradient_list
```

```python
initial_index = 50

gradient_list = get_grad_list(initial_point=initial_index)

plt.plot(gradient_list, ".", linewidth=2.3, color='red')
plt.plot(gradient_list, )
plt.show()
```

## [Problem 7] Visualization of the descent process for each initial value

```python
initial_points = [152,  146,  130, 120, 50]

for point in initial_points:
    gradient_list = get_grad_list(initial_point=point)
    plt.plot(gradient_list, label=str(point))
    plt.legend()
    plt.show()
```
