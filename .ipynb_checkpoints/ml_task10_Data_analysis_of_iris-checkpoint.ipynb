## [Problem 1] Data acquisition

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
X = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
y = pd.DataFrame(iris.target, columns=['Species'])
```

```python
X.head()
```

```python
y.head()
```

## [Problem 2] Combining data

```python
df = pd.concat([X, y], axis=1)
df.head()
```

## [Problem 6] Creating a diagram

```python
!pip install seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pie chart
plt.pie(df['Species'].value_counts(), labels=iris.target_names, autopct='%1.1f%%')
plt.title('Number of samples per label')
plt.show()
```

```python
# Box plot
for col in df.columns[:4]:
    sns.boxplot(x='Species', y=col, data=df)
    plt.title(f'Distribution of {col} for each label')
    plt.show()
```

```python
# Violin plot
for col in df.columns[:4]:
    sns.violinplot(x='Species', y=col, data=df)
    plt.title(f'Distribution of {col} for Each Label')
    plt.show()

# Explanation of Box plot vs Violin plot:
# Box plot shows summary statistics (median, quartiles), Violin plot shows the entire probability distribution of the data.
# Violin plot is more informative, but Box plot is more compact.
```

## [Problem 7] Confirming the relationship between features

```python
# Scatter plots
sns.scatterplot(x='sepal_length', y='sepal_width', hue='Species', data=df)
plt.title('Scatter Plot of sepal length vs sepal width')
plt.show()
```

```python
# Scatterplot matrix
sns.pairplot(df, hue='Species')
plt.suptitle('Scatterplot Matrix', y=1.02)
plt.show()
```

```python
# Correlation coefficient matrix
correlation_matrix = df.iloc[:, :4].corr()
correlation_matrix
```

```python
# Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Coefficient Matrix')
plt.show()
```
